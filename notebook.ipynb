```python
from sqlalchemy import inspect

assessment_instance =db.session.query(Assessment).all()

def make_df_dynamic(class_instance):
    """Convert SQLAlchemy class instance to DataFrame dynamically."""
    with test_app.app_context():
        mapper = inspect(class_instance)
        columns = [column.key for column in mapper.attrs]

        # Create a dictionary with column names and values
        data = {column: getattr(class_instance, column) for column in columns}

        # Create a DataFrame from the dictionary
        df = pd.DataFrame([data])

        return df

make_df_dynamic
```

```python
def make_df(x):
    with test_app.app_context():
        '''make class into dataframe for analysis'''
        data = db.session.query(x).all()
        df = pd.DataFrame(data)
        return df

def lit_scores_general():
    with test_app.app_context():
        assessments = make_df(Assessment)
        standards = make_df(AssessmentStandard)
        scores = make_df(AssessmentScores)

        lit_tests = assessments[['subject']] == 'Literacy'
        lit_scores = scores.where(lit_tests)
        lit_standards =standards.where(lit_tests)
        standard_lit_scores = lit_scores.merge(lit_standards, on = 'assessment_id')
        return standard_lit_scores.head()

lit_scores_general()
```
